```python
import pandas as pd
import numpy as np
import os
```

```python
DIR = '/home/tfir/KJS/FinalLog/Group_activity/'
flist = os.listdir(DIR)
flist.sort()
flist
```

```python
for s in flist:
    fname = DIR + s
    if s == 'log_20100508.csv':
        df = pd.read_csv(fname)
    else:
        df2 = pd.read_csv(fname)
        df = pd.merge(df, df2, on='actor_account', how='outer').fillna(0)
        df["total_party_time"] = df["total_party_time_x"] + df["total_party_time_y"]
        df["guild_join_count"] = df["guild_join_count_x"] + df["guild_join_count_y"]
        df["out_count"] = df["out_count_x"] + df["out_count_y"]
        df.drop(["total_party_time_x", "total_party_time_y",
                 "guild_join_count_x", "guild_join_count_y",
                 "out_count_x", "out_count_y"], axis=1, inplace=True)
df = df.sort_values(by='actor_account').reset_index(drop=True)
df
```

```python
df["average_party"] = df["total_party_time"] / df["out_count"]
df.drop(["out_count"], axis=1, inplace=True)
df
```

```python
label = pd.read_csv("/home/tfir/dataset/test_account.csv")
label.columns = ["actor_account"] 
label
```

```python
Group_activity = pd.merge(df, label, on='actor_account', how='outer').fillna(0)
Group_activity = Group_activity.sort_values(by="actor_account").reset_index(drop=True)
Group_activity
```

```python
Group_activity.to_csv("/home/tfir/KJS/FinalLog/Group_activity.csv", index=False, encoding='utf8')
```
