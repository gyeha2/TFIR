```
def bar_chart(feature):
    bot = alpha[alpha['class']==1][feature].value_counts()
    player = alpha[alpha['class']==0][feature].value_counts()
    df = pd.DataFrame([bot,player])
    df.index = ['bot','player']
    df.plot(kind='bar',stacked=True, figsize=(10,5))
```
