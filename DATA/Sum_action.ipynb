```python
import pandas as pd
import numpy as np
import os
```

```python
DIR = '/home/tfir/KJS/FinalLog/Player_action/'
flist = os.listdir(DIR)
flist.sort()
flist
```

```python
for s in flist:
    fname = DIR + s
    if s == 'log_20100508.csv':
        df1 = pd.read_csv(fname)
        max1 = df1['abyss']
        max1.index = df1['actor_account']
    else:
        df2 = pd.read_csv(fname)
        max2 = df2['abyss']
        max2.index = df2['actor_account']
        max1 = pd.merge(max1, max2, on='actor_account', how='outer').fillna(0).max(axis=1)
        max1.name = 'abyss'
max1 = max1.sort_index()
max1
```

```python
drop_columns = ["sit_count_x", "sit_count_y", 
                "exp_get_amount_x", "exp_get_amount_y", 
                "item_get_count_x", "item_get_count_y",
                "exp_repair_count_x", "exp_repair_count_y",
                "money_get_count_x", "money_get_count_y",
                "use_portal_count_x", "use_portal_count_y",
                "killed_by_pc_x", "killed_by_pc_y",
                "killed_by_npc_x", "killed_by_npc_y",
                "teleport_count_x", "teleport_count_y",
                "reborn_count_x", "reborn_count_y",
                "question_count_x", "question_count_y"]
for s in flist:
    fname = DIR + s
    if s == 'log_20100508.csv':
        df = pd.read_csv(fname)
        del df['abyss']
    else:
        df2 = pd.read_csv(fname)
        del df2['abyss']
        df = pd.merge(df, df2, on='actor_account', how='outer').fillna(0)
        df["sit_count"] = df["sit_count_x"] + df["sit_count_y"]
        df["exp_get_amount"] = df["exp_get_amount_x"] + df["exp_get_amount_y"]
        df["item_get_count"] = df["item_get_count_x"] + df["item_get_count_y"]
        df["exp_repair_count"] = df["exp_repair_count_x"] + df["exp_repair_count_y"]
        df["money_get_count"] = df["money_get_count_x"] + df["money_get_count_y"]
        df["use_portal_count"] = df["use_portal_count_x"] + df["use_portal_count_y"]
        df["killed_by_pc"] = df["killed_by_pc_x"] + df["killed_by_pc_y"]
        df["killed_by_npc"] = df["killed_by_npc_x"] + df["killed_by_npc_y"]
        df["teleport_count"] = df["teleport_count_x"] + df["teleport_count_y"]
        df["reborn_count"] = df["reborn_count_x"] + df["reborn_count_y"]
        df["question_count"] = df["question_count_x"] + df["question_count_y"]
        df.drop(drop_columns, axis=1, inplace=True)

df = df.sort_values(by='actor_account').reset_index(drop=True)
df
```

```python
Player_action = pd.merge(df, max1, on='actor_account', how='outer').fillna(0)
Player_action = Player_action[["actor_account", "sit_count", "exp_get_amount" ,"item_get_count", "exp_repair_count", "money_get_count", "abyss", "use_portal_count", "killed_by_pc", "killed_by_npc", "teleport_count", "reborn_count", "question_count"]]
Player_action
```

```python
Player_action = pd.merge(Player_action, label, on="actor_account", how="outer")
Player_action
```
